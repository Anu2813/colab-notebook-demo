```python
X = np.arange(1, 11)
x_scatter = np.random.rand(10)
y_scatter = np.random.rand(10)
favorite_foods = {
    "Upma": 3.50,
    "Dosa": 4.25,
    "Idly": 2.75
}
foods = list(favorite_foods.keys())
prices = list(favorite_foods.values())

normal_data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(10, 5))

axs[0, 0].plot(X, X/2)
axs[0, 0].set_title("X vs X/2")
axs[0, 0].set_xlabel("X")
axs[0, 0].set_ylabel("X/2")

axs[0, 1].scatter(x_scatter, y_scatter, color='green')
axs[0, 1].set_title("Random Scatter")
axs[0, 1].set_xlabel("X")
axs[0, 1].set_ylabel("Y")

axs[1, 0].bar(foods, prices, color='orange')
axs[1, 0].set_title("Favorite Foods Prices")
axs[1, 0].set_xlabel("Food")
axs[1, 0].set_ylabel("Price (USD)")

axs[1, 1].hist(normal_data, bins=30, edgecolor='black')
axs[1, 1].set_title("Normal Distribution Histogram")
axs[1, 1].set_xlabel("Value")
axs[1, 1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = '/content/drive/MyDrive/car-sales.csv'
car_sales = pd.read_csv(file_path)

# Step 1: Clean 'Price' column (remove $, commas and convert to float)
car_sales['Price'] = car_sales['Price'].replace('[\$,]', '', regex=True).astype(float)

# Step 2: Create 'Total Sales' column using cumulative sum
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Step 3: Create 'Sale Date' column with evenly spaced dates
car_sales['Sale Date'] = pd.date_range(start='2023-01-01', periods=len(car_sales), freq='D')

# Step 4: Plot the cleaned 'Price' column
car_sales['Price'].plot(title='Car Sale Prices Over Time', figsize=(10, 5))
plt.xlabel('Index')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()

# Optional: Plot cumulative sales
car_sales['Total Sales'].plot(title='Cumulative Total Sales', figsize=(10, 5), color='green')
plt.xlabel('Index')
plt.ylabel('Total Sales (USD)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/car-sales.csv'
car_sales = pd.read_csv(file_path)

car_sales['Price'] = car_sales['Price'].replace('[\$,]', '', regex=True)
car_sales['Price'] = car_sales['Price'].str.replace(r'\.\d{2}$', '', regex=True)
car_sales['Price'] = car_sales['Price'].astype(int)

car_sales['Total Sales'] = car_sales['Price'].cumsum()

car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(car_sales), freq='D')

plt.figure(figsize=(10, 5))
plt.plot(car_sales['Sale Date'], car_sales['Total Sales'], marker='o')

plt.title("Total Sales Over Time")
plt.xlabel("Sale Date")
plt.ylabel("Cumulative Sales (USD)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Code_Help/heart-disease.csv'
heart_disease = pd.read_csv(file_path)

over_50 = heart_disease[heart_disease['age'] > 50]

plt.figure(figsize=(10, 6))

scatter = plt.scatter(
    over_50['age'],
    over_50['chol'],
    c=over_50['target'],
    cmap='bwr',
    alpha=0.8
)

plt.title('Heart Disease Presence vs Age and Cholesterol (Age > 50)')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.colorbar(label='Heart Disease (0 = No, 1 = Yes)')

plt.axhline(y=over_50['chol'].mean(), color='green', linestyle='--', label='Mean Cholesterol')

plt.legend()
plt.tight_layout()
plt.show()
```
